```python
plt.figure(figsize=(14,10))
photo = io.imread('parrot.jpg')
plt.subplot(221)
plt.imshow(photo);
# nrows, ncols, plotnumber
plt.subplot(222)
# Flip horizontally
new_photo = photo.copy()
# use slicing operator
new_photo = new_photo[:,::-1]
plt.imshow(new_photo)
plt.subplot(223)
# Flip vertically
new_photo = photo.copy()
# use slicing operator
new_photo = new_photo[::-1]
plt.imshow(new_photo)
plt.subplot(224)
# Flip both
new_photo = photo.copy()
# use slicing operator
new_photo = new_photo[::-1,::-1]
plt.imshow(new_photo);
```

```python
# effects

plt.figure(figsize=(14,10))
photo = io.imread('parrot.jpg')
plt.subplot(221)
# nrows, ncols, plotnumber
plt.imshow(photo);
plt.subplot(222)
# remove blue color
new_photo = photo.copy()
# 0 - Red, 1- Green , 2 - Blue
new_photo[:,:,2] = 0
plt.imshow(new_photo);
plt.subplot(223)
# remove green color
new_photo = photo.copy()
# 0 - Red, 1- Green , 2 - Blue
new_photo[:,:,1] = 0
plt.imshow(new_photo);
plt.subplot(224)
# remove red color
new_photo = photo.copy()
# 0 - Red, 1- Green , 2 - Blue
new_photo[:,:,0] = 0
plt.imshow(new_photo);
```

```python
import numpy as np
from PIL import Image, ImageDraw
  
img=Image.open("parrot.jpg")
display(img)
  
height,width = img.size
lum_img = Image.new('L', [height,width] , 0)
  
draw = ImageDraw.Draw(lum_img)
draw.pieslice([(180,0), (height,width)], 0, 360, 
              fill = 255, outline = "white")
img_arr =np.array(img)
lum_img_arr =np.array(lum_img)
display(Image.fromarray(lum_img_arr))
final_img_arr = np.dstack((img_arr,lum_img_arr))
display(Image.fromarray(final_img_arr))
```
