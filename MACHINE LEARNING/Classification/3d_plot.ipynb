```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# fig= plt.figure()
# as = plt.axes(projection='3d')
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
X,Y=np.meshgrid(x,y)
Z=X+Y
z=X*np.exp(-X**2-Y**2)
```

```python
Z.shape
# 3d
```

```python
fig= plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='viridis')
# ax.contourf(X,Y,Z,cmap='viridis')
plt.show()
```

```python
fig= plt.figure()
ax = plt.axes(projection='3d')
# ax.plot_surface(X,Y,Z,cmap='viridis')
ax.contourf(X,Y,Z,cmap='viridis')
plt.show()
```
