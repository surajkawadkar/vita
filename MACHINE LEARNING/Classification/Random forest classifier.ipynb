```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
X=df.iloc[:,[0,1,2,3]].values
Y=df.iloc[:,-1].values
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,
                                               Y ,
                                               test_size= 0.2,
                                              random_state=123)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

# Random Foresr Classifire

```python
from  sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=5)  #n_estimator is number of tress
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
model.estimators_
```

```python
    from sklearn import tree
```

```python

cols=['sepel_length','sepel_width','petal_length','petal_width']
classes=['Setosa','versicolor','Verginica']
for i in range(5):
    fig, axes=plt.subplots(figsize=(12,12))
   
    tree.plot_tree(model.estimators_[i],
                   feature_names=cols,
              class_names=classes,
              filled=True,
                  ax=axes)
    
```

```python
model.feature_importances_
# we pass 4 column it shows how importance first 2 column are not that important can be seen
```

```python
plt.bar(x=cols,height=model.feature_importances_)
```

```python
plt.barh(y=cols,width=model.feature_importances_)
```
