```python
startups.fillna(startups.mean(),inplace=True)
# make permenant changes where na there and fill it with mean
```

```python
# Dealing with ordinary cell
```

```python
x=startups.iloc[:,[0,2,3]].values   #ignoring administrative column  
print(x.shape)

y=startups.iloc[:,-1].values
# profit col
print(y.shape)
```

```python
type(x)
```

# deals with string
## labeled encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x
```

```python
lab_en=LabelEncoder()
x[:,2]=lab_en.fit_transform(x[:,2])
```

```python
x
```

# one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(
        [('ohe', OneHotEncoder(categories='auto'),[2])],  #2 is a ratio
        remainder = 'passthrough'
        )
# list of tuples
#    List of (name, transformer, column(s)) tuples specifying the
#    transformer objects to be applied to subsets of the data.
```

```python
x=ct.fit_transform(x)
x
```


# split the data into model and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,
                                               y ,
                                               test_size= 0.2, 
                                               random_state=0 )
# random state is a seed value
# ypu can give any seed value
# to replicate the result , use same seed value
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model =LinearRegression()
model.fit(x_train,y_train)
print(model.coef_)
print(model.intercept_)
```

```python
model =LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
y_test
```

```python
r2_score(y_test,y_pred)
# truth value 1st, predicted value 2nd
# 87 % accuracy of the model
```
